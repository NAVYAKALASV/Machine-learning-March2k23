```python
#20BCE1858 S.V.NAVYA KALA- K MEANS WITH IN-BUILT FUNCTIONS
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import math
m1 = pd.read_csv("train.csv")
m2 = pd.read_csv("test.csv")
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_scaled = scaler.fit_transform(m1)
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=10, init='k-means++', max_iter=300, n_init=10, random_state=0)
pred_y = kmeans.fit_predict(df_scaled)
```

```python
from sklearn.metrics import silhouette_score

silhouette = silhouette_score(df_scaled, pred_y)
print("Silhouette score:", silhouette)

inertia = kmeans.inertia_
print("Inertia:", inertia)
```

```python
import numpy as np

inertias = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(df_scaled)
    inertias.append(kmeans.inertia_)
    
plt.plot(range(1, 11), inertias)
plt.title('Elbow plot')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```
